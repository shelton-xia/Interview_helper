```python
import pyautogui
import keyboard
from openai import OpenAI
import mss
import mss.tools
import requests

# Replace with your OpenAI API key
api_key = 'Your OpenAI API key'

def capture_screenshot_from_second_monitor():
    with mss.mss() as sct:
        # Get the dimensions of the second monitor
        monitors = sct.monitors
        if len(monitors) < 2:
            print("Second monitor not found!")
            return None
        
        second_monitor = monitors[2]
        screenshot = sct.grab(second_monitor)
        screenshot_path = 'screenshot.png'
        mss.tools.to_png(screenshot.rgb, screenshot.size, output=screenshot_path)
        return screenshot_path

def query_gpt4(image_path):
    # Convert image to base64
    import base64
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    
    
    client = OpenAI(api_key=api_key)
    response = client.chat.completions.create(
        model="gpt-4o",  # Ensure you're using the correct model
        messages=[
            {"role": "system", "content": "You are an software engineer expert with great coding skills in Python"},
            {
              "role": "user",
              "content": [
                {
                  "type": "text",
                  "text": """This is a screenshot of a coding problem. Please provide the code solution. Try to provide two solustions, first give the most optimized solution,then give a solution without importing packages. First come with the concise solution idea, time complexity and then exact codes.
                    Remember to - Write the optimized solution in **clean, well-commented code**."
                                - Explain each line or function to ensure full understanding."
                                - Run through a test case to validate correctness."""
                },
                {
                  "type": "image_url",
                  "image_url": {
                    "url": f"data:image/jpeg;base64,{encoded_string}"
                  }
                }
              ]
            }
        ],
    max_tokens=2000
    )
    
    return response.choices[0].message.content

def main():
    print("Press F3 to take a screenshot from the second monitor and get the solution from GPT-4.")
    while True:
        if keyboard.is_pressed('F2'):
            image_path = capture_screenshot_from_second_monitor()
            if image_path:
                print("Screenshot taken and saved as 'screenshot.png'")
                response = query_gpt4(image_path)
                print("GPT-4 Response:\n", response)
            break

if __name__ == "__main__":
    main()
```
